```python
import networkx as nx
import matplotlib.pyplot as plt

# Daftar simpul
nodes = ['a', 'b', 'c', 'd', 'e', 'f']

# Buat graf berarah lengkap
G = nx.DiGraph()
G.add_nodes_from(nodes)

# Tambahkan semua edge dari setiap pasangan simpul (tanpa self-loop)
for u in nodes:
    for v in nodes:
        if u != v:
            G.add_edge(u, v)

# Layout melingkar agar rapi
pos = nx.circular_layout(G)

# Gambar graf
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color='lightgreen', node_size=1000, font_size=14,
        arrows=True, edge_color='gray')
plt.title("Graf Berarah Lengkap (6 Simpul)")
plt.show()
```
